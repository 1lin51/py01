## 一.程式碼及註解

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.cross_validation import train_test_split
from sklearn import metrics, tree

url = "https://storage.googleapis.com/py_ds_basic/kaggle_titanic_train.csv"#資料網址
titanic_train = pd.read_csv(url)#用pandas的read讀入csv檔案

# 看看資料
print(titanic_train.head()) # 用hand()讀取前幾筆資料,目的在了解資料結構
print("---")
print(titanic_train.dtypes)# 用pandas的dtypes讀出資料型別

# 描述性統計
print(titanic_train.describe()) # 用pandas的describe做出資料的描述性統計,從統計中找尋資料的紀錄遺漏(欄位內無資料)以異常,發現891比資料中 Age(年齡)有遺漏(NaN)

#建立函數統計各個欄位的真實缺漏數(與描述性統計對應)
def justify_is_null(data):
    'Confirm the number of NaN for each variable'
    number_of_nan = [] #建立一個名為number_of_nan的list,準備接收各個欄位缺漏的計算
    for i in range(len(data.columns)): #建構for迴圈,遍歷每一個欄位
        number_of_nan.insert(i, sum(pd.isnull(data.ix[:,i]))) #這一行是三個程式功能的縮寫,第一是用pandas的isnull判斷欄位是否是空值,若是,則進行第二步的加總,然後再將加總值填入預設作為儲存用名為number_of_nan的list中
    return number_of_nan

print(justify_is_null(titanic_train))


#使用numpy的nanmedian計算中位數,利用中位數來填補 Age 變數的遺漏值,使用numpy的where填入age中的空欄位(isnull)
age_median = np.nanmedian(titanic_train.Age) # 使用要忽略 NaN
new_Age = np.where(titanic_train.Age.isnull(), age_median, titanic_train.Age)
titanic_train.Age = new_Age #將填補好的new_Age置入原來的titanic_train.Age
print(titanic_train.describe())# 用pandas的describe做出更新age後的新的資料的描述性統計

# 使用matplotlib.pyplot繪製長條圖,看看填補完後的情形
plt.hist(titanic_train.Age)
plt.show()


#填補登船港口遺漏值,以最多人登船處作為填補值
print(titanic_train.Embarked.value_counts()) # embarked 是登船處,用value_counts計算,顯示有3個登船處,其中以在S登船者為最多
new_Embarked = np.where(titanic_train.Embarked.isnull(), "S", titanic_train.Embarked)

#............以上是填補資料欄位空值的作業............


#............以下是利用sklearn的LabelEncoder,將特定資料轉化為標籤值...........

#from sklearn.preprocessing import LabelEncoder,建立 dummy variables
label_encoder = LabelEncoder() #建構一個LabelEncoder,LabelEncoder是對不連續的數字或文字予以特色標籤化
encoded_Sex = label_encoder.fit_transform(titanic_train.Sex) #將性別轉換成0與1
encoded_Embarked = label_encoder.fit_transform(new_Embarked) #將三個登船處分別予以編號標示為0,1,2
print(encoded_Sex)
print(encoded_Embarked)

#...........以下是為"利用sklearn"建立預測模型做準備............................

#Scikit-learn 完全使用矩陣運算。先用pd.DataFrame做出x,Y矩陣作為計算準備,公式要看老師的程式

titanic_X = pd.DataFrame([titanic_train.Pclass,
                          encoded_Sex,
                          new_Age,
                          titanic_train.SibSp,
                          titanic_train.Parch,
                          titanic_train.Fare,
                          encoded_Embarked
]).T #建構一個pd.DataFrame,將要作為模型的欄位值放入,作為x軸.....至於.T是啥還要查(應該是公式中的t值)

titanic_y = titanic_train.Survived #生還者的欄位作為Y軸

#from sklearn.cross_validation import train_test_split:引用train_test_split對資料進行切割
# 切分訓練(0.7)與測試資料(0.3)
train_X, test_X, train_y, test_y = train_test_split(titanic_X, titanic_y, test_size = 0.3)

#...........以下是"利用sklearn"建立預測模型............................

#from sklearn import metrics, tree
# 建立模型:引用sklearn的tree
decison_clf = tree.DecisionTreeClassifier(random_state = 87)
decison_clf.fit(train_X, train_y)

# 預測
test_y_predicted = decison_clf.predict(test_X)

# 準確率:引用sklearn的metrics
accuracy = metrics.accuracy_score(test_y, test_y_predicted)
print(accuracy)


#...........以下是將前述的預測模型套在新的資料中進行預測............................

url = "https://storage.googleapis.com/py_ds_basic/kaggle_titanic_test.csv"
to_submit = pd.read_csv(url)

# 繳交資料有一個觀測值的 Fare 是遺漏值
print(justify_is_null(to_submit))

# 跟訓練資料作一樣的整理
encoded_Sex_to_submit = label_encoder.fit_transform(to_submit.Sex)
encoded_Embarked_to_submit = label_encoder.fit_transform(to_submit.Embarked)
age_median = np.nanmedian(to_submit.Age) # 要忽略 NaN
imputed_Age = np.where(to_submit.Age.isnull(), age_median, to_submit.Age)
fare_median = np.nanmedian(to_submit.Fare) # 要忽略 NaN
imputed_Fare = np.where(to_submit.Fare.isnull(), fare_median, to_submit.Fare)

to_submit_X = pd.DataFrame([to_submit.Pclass,
                            encoded_Sex_to_submit,
                            imputed_Age,
                            to_submit.SibSp,
                            to_submit.Parch,
                            imputed_Fare,
                            encoded_Embarked_to_submit
]).T

# 預測
to_submit_y = decison_clf.predict(to_submit_X)
print(to_submit_y)

# 準備要上傳的檔案
to_submit_dict = {
    "PassengerId": to_submit["PassengerId"],
    "Survived": to_submit_y
}
to_submit_df = pd.DataFrame(to_submit_dict)

# 輸出成 csv
to_submit_df.to_csv("to_submit.csv", index = False)

```
